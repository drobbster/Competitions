### Data PreProcessing: XGBoost Model

```python
#### Preprocessing For XGBoost Model

def impute(X):
    
    # remove the colum that has ~20% null values, also ranks low on feature importance
#    X.drop(['ndvi_ne'], axis=1, inplace=True)
    
    # Filling the rest using linear interpolation
    X.interpolate(inplace=True)

def remove_outliers(df):
    return df[(np.abs(stats.zscore(df)) < 5).all(axis=1)]

#def mape(Y_test, Y_pred, epsilon = 1):
#    return np.mean(np.abs((Y_test - Y_pred + epsilon) / (Y_test + epsilon))) * 100

def extract_month(s):
    return int(s[5:7])

def city_indices(X):
    # city boolean encoding
    return X.city == 'sj'


def pre_process(X, trees = False):
    '''
    Extracts the month out of date and converts it to a one hot
    Standardizes the numerical features
    '''
    
    #Extracting month from the date
    months = X.week_start_date.apply(extract_month)

    # Removing the columns not required for classification
    X.drop(['city', 'year', 'weekofyear', 'week_start_date'], axis=1, inplace=True)

    # Standardizing the data
    if not trees:
        scaler = StandardScaler()
        X[X.columns] = scaler.fit_transform(X)

    # Month one hot features
    month_features = pd.get_dummies(months)
    X = X.join(month_features)

    # Alternatively use months as a discrete feature
    #X = X.join(months)

    return X

def seperate_cities_data(X, is_sj):
    '''
    Separating the data for the two cities in order to create a 
    different model for each
    '''

    # Separating the cities data
    X_sj = X.loc[is_sj]
    X_iq = X.loc[~is_sj]

    return X_sj, X_iq

def get_y_labels(X_sj, X_iq, y):

    y = y.total_cases
    y_sj = y.loc[X_sj.index]
    y_iq = y.loc[X_iq.index]

    return y_sj, y_iq

def split(X_sj, X_iq, y_sj, y_iq):

    # train and test split
    sj_split_data = train_test_split(X_sj, y_sj, shuffle = False)
    iq_split_data = train_test_split(X_iq, y_iq, shuffle = False)

    return sj_split_data, iq_split_data

def process(X, y = pd.Series(), train = True, trees = False):
    '''
    preprocesses the data
    

    Arguments:
    X - the data to be pre processed
    y - labels, optional
    train - if False, computes pre processing for test data set.
    trees - True if the model to be used is a tree based one, for eg. XGBoost
    '''

    is_sj = city_indices(X)
    if not trees:
        impute(X)
    X = pre_process(X, trees)

    #X = remove_outliers(X)
    X_sj, X_iq = seperate_cities_data(X, is_sj)
    if y.empty:
        return X_sj, X_iq

    y_sj, y_iq = get_y_labels(X_sj, X_iq, y)
    if not train:
        return X_sj, X_iq, y_sj, y_iq

    return split(X_sj, X_iq, y_sj, y_iq)
```

#### Generalized Model

A generalised model is nothing more than the combination of accuracy scores on different cities trained on different models!

```python
def general_model(clf, data):

    """
    returns the mean absolute error combined for both the cities.
    """
    (X_sj_train, X_sj_test, Y_sj_train, Y_sj_test), (X_iq_train, X_iq_test, Y_iq_train, Y_iq_test) = data
    
    clf.fit(X_sj_train, Y_sj_train, eval_metric = mean_absolute_error)
    Y_sj_pred = clf.predict(X_sj_test)

    clf.fit(X_iq_train, Y_iq_train, eval_metric = mean_absolute_error)
    Y_iq_pred = clf.predict(X_iq_test)
    
    Y_pred = np.concatenate([Y_sj_pred, Y_iq_pred])
    Y_pred = Y_pred.astype(int).clip(0)
    Y_test = Y_sj_test.append(Y_iq_test)

    return mean_absolute_error(Y_test, Y_pred)
```

#### Combined Tuned Model Performance

Combined accuracy scores of the two different tuned models for their respective cities.


```python
def combined_model_performance(clf1, clf2, data):

    """
    returns the mean absolute error combined for both the cities.
    """
    (X_sj_train, X_sj_test, Y_sj_train, Y_sj_test), (X_iq_train, X_iq_test, Y_iq_train, Y_iq_test) = data
    
    clf1.fit(X_sj_train, Y_sj_train, eval_metric = mean_absolute_error)
    Y_sj_pred = clf1.predict(X_sj_test)

    clf2.fit(X_iq_train, Y_iq_train, eval_metric = mean_absolute_error)
    Y_iq_pred = clf2.predict(X_iq_test)
    
    Y_pred = np.concatenate([Y_sj_pred, Y_iq_pred])
    Y_pred = Y_pred.astype(int).clip(0)
    Y_test = Y_sj_test.append(Y_iq_test)

    return mean_absolute_error(Y_test, Y_pred)
```

#### Data Preprocessing Functions For RNN Model

```python
## Reworked with edits from team input

def impute(X):
    
    # remove the colum that has ~20% null values, also ranks low on feature importance
#    X.drop(['ndvi_ne'], axis=1, inplace=True)
    
    # Filling the rest using linear interpolation
    X.interpolate(inplace=True)

def extract_month(s):
    return int(s[5:7])

def city_indices(X):
    # city boolean encoding
    return X.city == 'sj'


def pre_process(X, trees = False, scaler = None):
    '''
    Extracts the month out of date and converts it to a one hot
    Standardizes the numerical features
    '''
    
    #Extracting month from the date
    months = X.week_start_date.apply(extract_month)

    # Removing the columns not required for classification
    X.drop(['city', 'week_start_date'], axis=1, inplace=True)
    
    # Standardizing the data
    if not trees:
        if scaler is None:
            scaler = StandardScaler()
            scaler.fit(X)
        X[X.columns] = scaler.transform(X)   

    # Month one hot features
    month_features = pd.get_dummies(months)
    X = X.join(month_features)

    # Alternatively use months as a discrete feature
    #X = X.join(months)

    return X, scaler

def seperate_cities_data(X, is_sj):
    '''
    Separating the data for the two cities in order to create a 
    different model for each
    '''

    # Separating the cities data
    X_sj = X.loc[is_sj]
    X_iq = X.loc[~is_sj]

    return X_sj, X_iq

def get_y_labels(X_sj, X_iq, y):

    y = y.total_cases
    y_sj = y.loc[X_sj.index]
    y_iq = y.loc[X_iq.index]

    return y_sj, y_iq

def split(X_sj, X_iq, y_sj, y_iq):

    # train and test split
    sj_split_data = train_test_split(X_sj, y_sj, shuffle = False)
    iq_split_data = train_test_split(X_iq, y_iq, shuffle = False)

    return sj_split_data, iq_split_data

def process(X, y = pd.Series(), train = True, trees = False, scaler = None):
    '''
    preprocesses the data
    

    Arguments:
    X - the data to be pre processed
    y - labels, optional
    train - if False, computes pre processing for test data set.
    trees - True if the model to be used is a tree based one, for eg. XGBoost
    scaler - contains the scaler transformation information for use on the hold out set
    '''

    is_sj = city_indices(X)
    if not trees:
        impute(X)
    X, scaler = pre_process(X, trees, scaler)

    #X = remove_outliers(X)
    X_sj, X_iq = seperate_cities_data(X, is_sj)
    if y.empty:
        return X_sj, X_iq, scaler

    y_sj, y_iq = get_y_labels(X_sj, X_iq, y)
    if not train:
        return X_sj, X_iq, y_sj, y_iq

    return split(X_sj, X_iq, y_sj, y_iq), scaler
```

#### Batch Generator Functions

Here is the original data generator from the Chollet book. It yields a tuple `(samples, targets)` where `samples` is one batch of input data and 
`targets` is the corresponding array of target temperatures. It takes the following arguments:

* `data`: The original array of floating point data, which we just normalized in the code snippet above.
* `lookback`: How many timesteps back should our input data go.
* `delay`: How many timesteps in the future should our target be.
* `min_index` and `max_index`: Indices in the `data` array that delimit which timesteps to draw from. This is useful for keeping a segment 
of the data for validation and another one for testing.
* `shuffle`: Whether to shuffle our samples or draw them in chronological order.
* `batch_size`: The number of samples per batch.
* `step`: The period, in timesteps, at which we sample data. We will set it 6 in order to draw one data point every hour.

#### Original Source Code Generator

```python
def generator(data, lookback, delay, min_index, max_index,
              shuffle=False, batch_size=1, step=1):
    if max_index is None:
        max_index = len(data) - delay - 1
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows),
                           lookback // step,
                           data.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = data[indices]
            targets[j] = data[rows[j] + delay][1]
        yield samples, targets
```

### Generator 2:
###### Modified For Use W/ 3D Tensor Shaped Data 

```python
def generator2(X, y, lookback, delay, min_index, max_index,
              shuffle=False, batch_size=1, step=1):
    if max_index is None:
        max_index = len(X[0]) - delay - 1
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows),
                           lookback // step,
                           X.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = X[0][indices]
            targets[j] = y[0][rows[j] + delay]
        yield samples, targets
```

```python
def reverse_generator2(X, y, lookback, delay, min_index, max_index,
              shuffle=False, batch_size=1, step=1):
    if max_index is None:
        max_index = len(X[0]) - delay - 1
    i = min_index + lookback

    while 1:
        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows),
                           lookback // step,
                           X.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = X[0][::-1][indices]
            targets[j] = y[0][::-1][rows[j] + delay]
        yield samples, targets
        
#[x[::-1]
```

### Generator 3: 
##### Used with holdout data for predicting cases with the predict_generator method

```python

def generator3(X, y, lookback, delay, min_index, max_index,
              shuffle=False, batch_size=1, step=1):
    if max_index is None:
        max_index = len(X[0]) - delay - 1
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows),
                           lookback // step,
                           X.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = X[0][indices]
#            targets[j] = y[0][rows[j] + delay]
        yield samples #, targets
```

### Prediction using Trained XGBoost Regressor Model

```python
#### Preprocessing For XGBoost Model

def impute(X):
    
    # remove the colum that has ~20% null values, also ranks low on feature importance
#    X.drop(['ndvi_ne'], axis=1, inplace=True)
    
    # Filling the rest using linear interpolation
    X.interpolate(inplace=True)

def remove_outliers(df):
    return df[(np.abs(stats.zscore(df)) < 5).all(axis=1)]

#def mape(Y_test, Y_pred, epsilon = 1):
#    return np.mean(np.abs((Y_test - Y_pred + epsilon) / (Y_test + epsilon))) * 100

def extract_month(s):
    return int(s[5:7])

def city_indices(X):
    # city boolean encoding
    return X.city == 'sj'


def pre_process(X, trees = False):
    '''
    Extracts the month out of date and converts it to a one hot
    Standardizes the numerical features
    '''
    
    #Extracting month from the date
    months = X.week_start_date.apply(extract_month)

    # Removing the columns not required for classification
    X.drop(['city', 'year', 'weekofyear', 'week_start_date'], axis=1, inplace=True)

    # Standardizing the data
    if not trees:
        scaler = StandardScaler()
        X[X.columns] = scaler.fit_transform(X)

    # Month one hot features
    month_features = pd.get_dummies(months)
    X = X.join(month_features)

    # Alternatively use months as a discrete feature
    #X = X.join(months)

    return X

def seperate_cities_data(X, is_sj):
    '''
    Separating the data for the two cities in order to create a 
    different model for each
    '''

    # Separating the cities data
    X_sj = X.loc[is_sj]
    X_iq = X.loc[~is_sj]

    return X_sj, X_iq

def get_y_labels(X_sj, X_iq, y):

    y = y.total_cases
    y_sj = y.loc[X_sj.index]
    y_iq = y.loc[X_iq.index]

    return y_sj, y_iq

def split(X_sj, X_iq, y_sj, y_iq):

    # train and test split
    sj_split_data = train_test_split(X_sj, y_sj, shuffle = False)
    iq_split_data = train_test_split(X_iq, y_iq, shuffle = False)

    return sj_split_data, iq_split_data

def process(X, y = pd.Series(), train = True, trees = False):
    '''
    preprocesses the data
    

    Arguments:
    X - the data to be pre processed
    y - labels, optional
    train - if False, computes pre processing for test data set.
    trees - True if the model to be used is a tree based one, for eg. XGBoost
    '''

    is_sj = city_indices(X)
    if not trees:
        impute(X)
    X = pre_process(X, trees)

    #X = remove_outliers(X)
    X_sj, X_iq = seperate_cities_data(X, is_sj)
    if y.empty:
        return X_sj, X_iq

    y_sj, y_iq = get_y_labels(X_sj, X_iq, y)
    if not train:
        return X_sj, X_iq, y_sj, y_iq

    return split(X_sj, X_iq, y_sj, y_iq)
```
